```python
from pyspark.sql import SparkSession
import pandas as pd
```

```python
spark = SparkSession \
        .builder \
        .config("spark.driver.memory", "4g") \
        .appName("Repartition Example") \
        .config('spark.driver.maxResultSize', '4g') \
        .config('spark.sql.execution.arrow.pyspark.enabled', 'true') \
        .config('spark.sql.execution.arrow.pyspark.fallback.enabled', 'true') \
        .getOrCreate()
```

```python
# TODO Path to your file
df = spark.read.csv("c:/temp/parking_violation.csv", header=True)
#df = spark.read.format('csv').load("c:/temp/parking_violation.csv")
```

```python
#print(df.count())
#print(df.printSchema)

pd.set_option('max_colwidth', 200)
df_pd = df.limit(2000000).toPandas()
df_pd.to_csv("2000000.csv", header=True, sep =';')
```
